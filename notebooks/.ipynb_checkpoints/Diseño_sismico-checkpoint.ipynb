```python
# === FOLD 3D: Cálculo + Gráficos + (Opcional) Georreferencia ===
# Probado en Colab y local (Python 3.10+). No usa seaborn.

import math
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from pathlib import Path

# (Opcional) CSV con pandas
WRITE_CSV = True
try:
    import pandas as pd
except Exception:
    WRITE_CSV = False

# ----------------------------
# Rutas (Colab usa /content/)
# ----------------------------
BASE_DIR = Path("/content") if Path("/content").exists() else Path(".")
OUTDIR = BASE_DIR / "salidas_fold"
OUTDIR.mkdir(parents=True, exist_ok=True)

# Fuente para evitar warnings en Matplotlib
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'
```

```python
# ==================== PARÁMETROS DEL DISEÑO ====================
length_m = 14500.0     # m (inline = eje y)
width_m  = 7200.0      # m (crossline = eje x)

# Parámetros geofísicos
v   = 2300.0   # m/s
fmax = 90.0    # Hz
buz  = 10.0    # grados

# Profundidades
Zsl   = 500.0    # m (shallow layer)
Zob   = 2500.0   # m (objetivo)
Zdeep = Zob

# Config patch
n_half = 8       # nº de líneas REC por mitad del patch (total = 2*n_half+1)
```

```python
# ==================== CÁLCULOS SÍSMICOS BASE ====================
RI = v / (3.0 * fmax * math.sin(math.radians(buz)))  # espaciamiento a lo largo (REC)
SI = RI                                              # espaciamiento a lo largo (SRC, asumido igual)

bin_inline = RI / 2.0
bin_xline  = SI / 2.0
CMP_Bin    = bin_inline * bin_xline

near_offset = Zsl / 1.2
RLI = near_offset / math.sqrt(2.0)  # intervalo entre líneas de receptores
SLI = RLI                            # intervalo entre líneas de fuentes

Xmax = Zdeep

num = (2.0 * Xmax)**2 - (n_half * RLI)**2
if num <= 0:
    raise ValueError("Con el n_half elegido y RLI, no se alcanza el Xmax deseado.")
m_half = math.ceil(math.sqrt(num) / SLI)  # nº de líneas SRC por mitad del patch
```

```python
# ==================== UTILIDADES GEOMÉTRICAS ====================
def make_receiver_grid(length_m, width_m, RI, RLI):
    """
    Receptores en líneas paralelas al eje y (inline):
      - Líneas en x = k*RLI   (crossline)
      - Estaciones en y = i*RI (inline)
    Retorna: rec (N,2) en (x,y), nº de líneas y nº de receptores por línea.
    """
    n_rlines = int(np.floor(width_m / RLI)) + 1
    n_rperline = int(np.floor(length_m / RI)) + 1
    x_rlines = np.arange(n_rlines) * RLI
    y_r = np.arange(n_rperline) * RI
    XX = np.repeat(x_rlines, n_rperline)
    YY = np.tile(y_r, n_rlines)
    return np.column_stack([XX, YY]), n_rlines, n_rperline

def make_source_grid(length_m, width_m, SI, SLI):
    """
    Fuentes en líneas paralelas al eje x (crossline):
      - Líneas en y = j*SLI   (inline)
      - Fuentes en x = l*SI   (crossline)
    """
    n_slines = int(np.floor(length_m / SLI)) + 1
    n_sperline = int(np.floor(width_m / SI)) + 1
    y_slines = np.arange(n_slines) * SLI
    x_s = np.arange(n_sperline) * SI
    XX = np.tile(x_s, n_slines)
    YY = np.repeat(y_slines, n_sperline)
    return np.column_stack([XX, YY]), n_slines, n_sperline

def compute_pairs_midpoints(rec, src, n_rlines, n_rperline,
                            RLI, SLI, n_half, m_half, Xmax):
    """
    Reglas de selección:
      - |x_src - x_rec| <= n_half * RLI   (ventana crossline)
      - |y_src - y_rec| <= m_half * SLI   (ventana inline)
      - sqrt((x_s-x_r)^2 + (y_s-y_r)^2) <= Xmax  (offset máximo)
    Devuelve midpoints (x,y) en coordenadas de survey (m).
    """
    midpoints = []
    for x_s, y_s in src:
        k_center = int(round(x_s / RLI))
        k_min = max(0, k_center - n_half)
        k_max = min(n_rlines - 1, k_center + n_half)
        for k in range(k_min, k_max + 1):
            start = k * n_rperline
            end   = start + n_rperline
            line_rec = rec[start:end]
            mask_y = np.abs(line_rec[:,1] - y_s) <= (m_half * SLI)
            if not np.any(mask_y):
                continue
            cand = line_rec[mask_y]
            dx = cand[:,0] - x_s
            dy = cand[:,1] - y_s
            dist = np.hypot(dx, dy)
            m = dist <= Xmax
            if not np.any(m):
                continue
            mids = 0.5 * (cand[m] + np.array([x_s, y_s]))
            midpoints.append(mids)
    if not midpoints:
        return np.empty((0,2), dtype=np.float64)
    return np.vstack(midpoints)

def fold_map_from_pairs(mid_xy, n_bins_x, n_bins_y, bin_x, bin_y):
    i = np.floor_divide(mid_xy[:,0].astype(np.float64), bin_x).astype(np.int32)  # col (x)
    j = np.floor_divide(mid_xy[:,1].astype(np.float64), bin_y).astype(np.int32)  # fila (y)
    valid = (i >= 0) & (i < n_bins_x) & (j >= 0) & (j < n_bins_y)
    i = i[valid]; j = j[valid]
    fold = np.zeros((n_bins_y, n_bins_x), dtype=np.int32)  # filas=y (inline), cols=x (crossline)
    lin = j * n_bins_x + i
    counts = Counter(lin.tolist())
    for lin_idx, c in counts.items():
        jj = lin_idx // n_bins_x
        ii = lin_idx %  n_bins_x
        fold[jj, ii] = c
    return fold

def fold_stats(fold):
    vals = fold.ravel()
    covered = vals > 0
    if not covered.any():
        return {"min":0,"max":0,"mean":0.0,"median":0.0,"p10":0.0,"p50":0.0,"p90":0.0,"mode":0,"coverage_ratio":0.0}
    c = Counter(vals[covered].tolist())
    return {
        "min": int(vals[covered].min()),
        "max": int(vals[covered].max()),
        "mean": float(vals[covered].mean()),
        "median": float(np.median(vals[covered])),
        "p10": float(np.percentile(vals[covered], 10)),
        "p50": float(np.percentile(vals[covered], 50)),
        "p90": float(np.percentile(vals[covered], 90)),
        "mode": int(c.most_common(1)[0][0]),
        "coverage_ratio": float(covered.mean())
    }

def crop_core(fold, bin_xline, bin_inline, RLI, SLI, n_half, m_half):
    margin_x_bins = int(np.ceil((n_half*RLI)/bin_xline))
    margin_y_bins = int(np.ceil((m_half*SLI)/bin_inline))
    core = fold[margin_y_bins: fold.shape[0]-margin_y_bins,
                margin_x_bins: fold.shape[1]-margin_x_bins]
    return core

# ==================== CÁLCULO PRINCIPAL ====================
rec, n_rlines, n_rperline = make_receiver_grid(length_m, width_m, RI, RLI)
src, n_slines, n_sperline = make_source_grid(length_m, width_m, SI, SLI)
mid_xy = compute_pairs_midpoints(rec, src, n_rlines, n_rperline, RLI, SLI, n_half, m_half, Xmax)

# Malla CMP y fold
n_bins_x = int(np.floor(width_m  / bin_xline)) + 1
n_bins_y = int(np.floor(length_m / bin_inline)) + 1
fold = fold_map_from_pairs(mid_xy, n_bins_x, n_bins_y, bin_xline, bin_inline)

# Estadísticas
stats_all  = fold_stats(fold)
core       = crop_core(fold, bin_xline, bin_inline, RLI, SLI, n_half, m_half)
stats_core = fold_stats(core) if core.size > 0 else stats_all

# ==================== GUARDAR PAQUETE ====================
np.savez_compressed(
    OUTDIR / "fold_package.npz",
    fold=fold,
    mid_xy=mid_xy,
    n_bins_x=n_bins_x, n_bins_y=n_bins_y,
    bin_xline=bin_xline, bin_inline=bin_inline,
    width_m=width_m, length_m=length_m,
    RI=RI, SI=SI, RLI=RLI, SLI=SLI,
    CMP_Bin=CMP_Bin, Xmax=Xmax,
    n_half=n_half, m_half=m_half
)

if WRITE_CSV:
    pd.DataFrame([{"region":"all", **stats_all}, {"region":"core", **stats_core}]).to_csv(OUTDIR/"fold_stats.csv", index=False)

print("Paquete guardado en:", (OUTDIR / "fold_package.npz").resolve())
print("Stats ALL :", stats_all)
print("Stats CORE:", stats_core)

# ==================== GRÁFICOS ====================
# 1) Mapa de fold
plt.figure(figsize=(8,6))
extent = [0, width_m, 0, length_m]  # x=crossline, y=inline
plt.imshow(fold, origin="lower", extent=extent, aspect="auto")
plt.xlabel("Crossline (m)")
plt.ylabel("Inline (m)")
plt.title("Mapa de Fold (CMP por bin)")
plt.colorbar(label="Fold")
plt.savefig(OUTDIR/"fold_map.png", dpi=200, bbox_inches="tight")
plt.show()

# 2) Histograma (solo bins con fold>0)
vals = fold[fold > 0].ravel()
plt.figure(figsize=(7,5))
if vals.size > 0:
    plt.hist(vals, bins=40)
plt.xlabel("Fold")
plt.ylabel("Frecuencia de bins")
plt.title("Distribución de Fold (bins con cobertura)")
plt.savefig(OUTDIR/"fold_hist.png", dpi=200, bbox_inches="tight")
plt.show()

# 3) Cobertura acumulada
plt.figure(figsize=(7,5))
if vals.size > 0:
    s = np.sort(vals)
    thr = np.unique(s)
    frac = [np.mean(s >= t) for t in thr]
    plt.plot(thr, frac)
plt.xlabel("Umbral de Fold")
plt.ylabel("Fracción de bins (≥ umbral)")
plt.title("Cobertura acumulada vs. umbral de fold")
plt.savefig(OUTDIR/"fold_coverage_curve.png", dpi=200, bbox_inches="tight")
plt.show()

print("Gráficos guardados en:", OUTDIR.resolve())
```

```python
# ==================== GEOREFERENCIA → GeoTIFF ====================
WRITE_GEOTIFF = True
from pyproj import CRS, Transformer
import rasterio
from rasterio.crs import CRS as CRS_rio
from rasterio.transform import Affine


if WRITE_GEOTIFF:
    # ---- Define ORIGEN y AZIMUT del survey ----
    # Origen geográfico (lat/lon) y CRS UTM de salida:
    lat0 = 4.3000
    lon0 = -72.0800
    def utm_epsg_from_lonlat(lon, lat):
        zone = int((lon + 180) // 6) + 1
        return (32600 if lat >= 0 else 32700) + zone
    epsg_out = utm_epsg_from_lonlat(lon0, lat0)  # elige UTM según el origen
    azimuth_inline_deg = 0.0  # 0°=inline hacia el norte; cámbialo si tu inline está rotado

    # Transforma origen (lon/lat) a UTM (metros)
    crs_geo = CRS.from_epsg(4326)
    crs_utm = CRS.from_epsg(epsg_out)
    to_utm = Transformer.from_crs(crs_geo, crs_utm, always_xy=True)
    E0, N0 = to_utm.transform(lon0, lat0)  # origen en metros (Easting, Northing)

    # Construye la geotransform (afín) con rotación
    # Definimos el raster con fila 0 en la parte NORTE (convención GIS):
    # → por eso usamos una versión "flipped" del fold para que la fila 0 sea el borde norte.
    fold_gdal = np.flipud(fold).astype(np.float32)

    # Vectores de paso por columna y fila (en metros, en UTM), con rotación por azimut
    theta = math.radians(azimuth_inline_deg)
    # Matriz que mapea [x_crossline; y_inline] → [E; N]
    R = np.array([[ math.sin(theta),  math.cos(theta)],
                  [-math.cos(theta),  math.sin(theta)]], dtype=float)

    # Paso al avanzar 1 columna (bin en crossline, hacia la derecha en imagen):
    v_col = R @ np.array([bin_xline, 0.0])
    # Paso al avanzar 1 fila hacia ABAJO (norte→sur): usar -bin_inline (para norte-arriba)
    v_row = R @ np.array([0.0, -bin_inline])

    # Coordenadas UTM del vértice superior-izquierdo (esquina) del raster:
    # Tomamos la esquina NW del grid del survey (x=0, y=length_m)
    corner_nw = R @ np.array([0.0, length_m]) + np.array([E0, N0])

    # Affine (a,b,c,d,e,f): x = a*col + b*row + c ; y = d*col + e*row + f
    a, d = v_col[0], v_col[1]   # contribución de la columna
    b, e = v_row[0], v_row[1]   # contribución de la fila (nota: suele dar e<0 si norte-arriba)
    c, f = corner_nw[0], corner_nw[1]  # esquina superior-izquierda

    transform = Affine(a, b, c, d, e, f)

    # Escribimos GeoTIFF (1 banda, tipo float32)
    tif_path = OUTDIR / "fold_map.tif"
    with rasterio.open(
        tif_path,
        "w",
        driver="GTiff",
        height=fold_gdal.shape[0],
        width=fold_gdal.shape[1],
        count=1,
        dtype=fold_gdal.dtype,
        crs=CRS_rio.from_epsg(epsg_out),
        transform=transform,
        nodata=0.0,
        compress="lzw"
    ) as dst:
        dst.write(fold_gdal, 1)

    print("GeoTIFF escrito:", tif_path.resolve())
    print(f"CRS EPSG:{epsg_out}  |  Azimut inline={azimuth_inline_deg}°")
```
